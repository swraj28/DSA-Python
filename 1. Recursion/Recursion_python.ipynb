1.Binary Exponentitation: time:-O(log(n)) and space:-O(log(n))-->because of the call stack

```python
def power(a,n):
    if(n==0):
        return 1
    
    if(n%2==0):
        return power(a,n//2)*power(a,n//2)
    
    return power(a,n//2)*power(a,n//2)*a

a=int(input())
b=int(input())

print(power(a,b))
```

```python
#time -->o(log(n))
#space -->o(1)
def power_iterative(a,n):
    
    ans=1
    while(n):
        if(n&1):
            ans*=a
        a*=a
        n>>=1
    return ans


a=int(input())
b=int(input())

print(power_iterative(a,b))
```

2.Fibonacci Number

```python
def fibonacci(n):
    if(n==0 or n==1):
        return n
    
    return fibonacci(n-1)+fibonacci(n-2)

n=int(input())
print(fibonacci(n))
```

Check if the list is sorted or not

```python
#  Tail call optimized

def is_sorted(arr,n,si):
    if(si>=(n-1)):
        return True
    if(arr[si]>arr[si+1]):
        return False
    
    return is_sorted(arr,n,si+1)

n=int(input())
arr=[]
for i in range(n):
    x=int(input())
    arr.append(x)

print(is_sorted(arr,n,0))
```

Check if a number is present in List or not

```python
def is_present(arr,n,si,val):
    if(si>=n):
        return False
    
    if(arr[si]==val):
        return True
    
    return is_present(arr,n,si+1,val)

n=int(input())
arr=[]
for i in range(n):
    x=int(input())
    arr.append(x)

val=int(input())

if(is_present(arr,n,0,val)):
    print("The value is present in the array")
else:
    print("The Value is not Present")
```

First Index of Number

```python
def first_idx(arr,n,si,val):
    if(si>=n):
        return -1
    
    if(arr[si]==val):
        return si
    
    return first_idx(arr,n,si+1,val)

n=int(input())
arr=[]
for i in range(n):
    x=int(input())
    arr.append(x)

val=int(input())

print(first_idx(arr,n,0,val))
```

Last Index Of Number

```python
def last_idx(arr,n,si,val):
    if(si==n):
        return -1
    
    idx_found=last_idx(arr,n,si+1,val)
    
    if(idx_found==-1):
        if(arr[si]==val):
            return si;
        else:
            return -1
        
    return idx_found


n=int(input())
arr=[]
for i in range(n):
    x=int(input())
    arr.append(x)

val=int(input())

print(last_idx(arr,n,0,val))
```

Given a string, compute recursively a new string where all 'b' chars have been replaced by 'a'.

```python
# String Object doesnot support item assignment as it hapeens in c and c++. To make a string 
# work the same way as it does in c++, then we can use a list in place of string.
def recur(s,b,a):
    if(len(s)==0):
        return s
    
    small_ans=recur(s[1:],a,b)
    
    if(s[0]==b):
        return a+small_ans
    else:
        return b+small_ans
    

# s=input()
s="abcabadebc"
recur(s,"b","a")
```

Given a string, compute recursively a new string where all 'x' chars have been removed.

```python
def recur(s,ans):
    if(len(s)==0):
        return ans
    if(s[0]!="x"):
        ans+=s[0]
    
    return recur(s[1:],ans)

s="abcxdxeda"
ans=""
recur(s,ans)
# print(ans)
```

Binary Search Using Recursion

```python
def binary_search(lst,st,end,val):
    if(st>end):
        return -1
    
    # Floor division //
    # Float division /
    mid=(st+(end-st)//2)
    
    if(lst[mid]==val):
        return mid
    
    if(lst[mid]<val):
        return binary_search(lst,mid+1,end,val)
    else:
        return binary_search(lst,st,mid-1,val)
    

lst=[1,2,3,4,5,6]
lst.sort() # to sort a list in python 
n=len(lst)

binary_search(lst,0,n-1,1)
    
```

```python
def merge(lst,st,end):
    arr=[]
    
    mid=(st+(end-st)//2)
    
    i=0
    j=mid+1
    while(i<=mid and j<=end):
        if(lst[i]<=lst[j]):
            arr.append(lst[i])
            i+=1
        else:
            arr.append(lst[j])
            j+=1
    
    while(i<=mid):
        arr.append(lst[i])
        i+=1
    
    while(j<=end):
        arr.append(lst[j])
        j+=1
    
    pos=0
    for i in range (st,end+1):
        lst[st]=arr[pos]
        pos+=1

def merge_sort(lst,st,end):
    if(st>=end):
        return
    
    mid=(st+(end-st)//2)
    
    merge_sort(lst,st,mid)
    merge_sort(lst,mid+1,end)
    
    merge(lst,st,end)

lst=[4,3,2,1,5]
n=len(lst)
merge_sort(lst,0,n-1)
lst
```
